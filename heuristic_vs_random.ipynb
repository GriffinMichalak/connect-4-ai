```python
from testconnect4 import TestConnect4Game
from models import HeuristicAI, RandomAI
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
```

```python
NUM_GAMES = 5000
```

```python
print("Starting Heuristic vs Random game...")

player_heuristic = HeuristicAI(player_id=1)
player1_wins = 0
player1_moves = []
player1_cols = []

player_random = RandomAI(player_id=2)
player2_wins = 0
player2_moves = []
player2_cols = []

game = TestConnect4Game(player1_ai=player_heuristic, player2_ai=player_random)
for g in range(NUM_GAMES):
    winner = game.run()
    winner_name = ""
    if winner == 1:
        print(f"Game {g}: Red (1) won in {game.player1_moves} moves")
        player1_moves.append(game.player1_moves)
        player1_wins += 1
    elif winner == 2:
        print(f"Game {g}: Yellow (2) won in {game.player2_moves} moves")
        player2_moves.append(player2_moves)
        player2_wins += 1
    else:
        print("Draw!")
        player2_wins += 1
    player1_cols.append(game.player1_cols)
    player2_cols.append(game.player2_cols)
    game.reset_game()

print()
print("========================= Stats =========================")
print(f"Red (1) won {player1_wins} times ({player1_wins / NUM_GAMES * 100}%)")
print(f"Yellow (2) won {player2_wins} times ({player2_wins / NUM_GAMES * 100}%)")
print(f"Red (1) Placements: {np.concatenate(player1_cols)}")
print(f"Yellow (2) Placements: {np.concatenate(player2_cols)}")
```

## Red (Player 1) Placement Heatmap

```python
placements = np.concatenate(player1_cols)

cols = np.array([0, 1, 2, 3, 4, 5, 6])

item_counts = Counter(placements)
counts = [item_counts.get(col, 0) for col in cols]

plt.figure(figsize=(8, 6))
plt.bar(cols, counts, color='red')

plt.xlabel('Column (0-6)')
plt.ylabel('Red (Player 1) Placement Count')
plt.title('HeuristicAI Placement Heatmap Against RandomAI')

plt.xticks(ticks=cols)
plt.tight_layout()
plt.show()
```

## Yellow (Player 2) Placement Heatmap

```python
placements = np.concatenate(player2_cols)

cols = np.array([0, 1, 2, 3, 4, 5, 6])

item_counts = Counter(placements)
counts = [item_counts.get(col, 0) for col in cols]

plt.figure(figsize=(8, 6))
plt.bar(cols, counts, color='yellow')

plt.xlabel('Column (0-6)')
plt.ylabel('Yellow (Player 2) Placement Count')
plt.title('Yellow (Player 2) Placement Heatmap')

plt.xticks(ticks=cols)
plt.tight_layout()
plt.show()
```
